# Occupation

### Paso 1. Importar librerias

```python
import numpy as np
import pandas as pd
```

### Paso 2. Importar el conjunto de datos de esta [dirección](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
```

### Paso 3. Asígnalo a una variable llamada usuarios.

```python
usuarios = pd.read_csv(url, sep = "|")
usuarios.head(4)
```

### Paso 4. Descubre cuál es la edad media por ocupación

```python
usuarios.groupby("occupation",as_index=False)["age"].mean()
```

### Paso 5. Descubra la proporción de Hombres por ocupación y ordénela de mayor a menor

```python
ocupacion_hombres = usuarios[usuarios["gender"]=="M"].groupby("occupation").size() / usuarios.groupby("occupation").size()

ordenar = ocupacion_hombres.sort_values(ascending=False)

ordenar.reset_index(name='ocupacion_hombres')
```

### Paso 6. Para cada ocupación, calcule las edades mínima y máxima

```python
usuarios.groupby("occupation",as_index=False)["age"].max()
```

```python
usuarios.groupby("occupation",as_index=False)["age"].min()
```

### Paso 7. Para cada combinación de ocupación y sexo, calcule la edad media

```python
ocupacion = usuarios.groupby("occupation",as_index=False)["age"].mean()
genero = usuarios.groupby("gender",as_index=False)["age"].mean()

print(ocupacion)
print(genero)
```

### Paso 8.  Para cada ocupación, presente el porcentaje de mujeres y hombres

```python
porcentaje = usuarios.groupby(["occupation", "gender"]).size().unstack(fill_value=0)

genero = porcentaje.div(porcentaje.sum(axis=1), axis=0) * 100

genero
```
