### Paso 7. Vea sólo las columnas Equipo, Tarjetas Amarillas y Tarjetas Rojas y asígnelas a un marco de datos llamado disciplina

```python
col = euro12[["Team","Yellow Cards","Red Cards"]]

disciplina = col

disciplina
```

### Paso 8. Ordena los equipos por tarjetas rojas y luego por tarjetas amarillas.

```python
roja = disciplina[["Team","Red Cards"]].sort_values(by=["Red Cards","Team"]) 
roja

amarilla = disciplina[["Team","Yellow Cards"]].sort_values(by=["Yellow Cards","Team"]) 
amarilla

roja.reset_index(drop=True, inplace=True)
amarilla.reset_index(drop=True, inplace=True)

resultado = pd.concat([roja, amarilla], axis=1)

resultado
```
