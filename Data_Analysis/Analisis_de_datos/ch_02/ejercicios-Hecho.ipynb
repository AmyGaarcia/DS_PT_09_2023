> 2. Encuentra el porcentaje de terremotos en Indonesia que fueron acompañados de tsunamis.

```python
indonesia = terremotos[(terremotos["type"]== "earthquake") & (terremotos["country"]=="Indonesia")]

total_terremotos = len(indonesia)

tsunamis = indonesia[indonesia["tsunami"] == True]

total_tsunamis = len(tsunamis)

if total_tsunamis > 0:
    porcentaje = (total_tsunamis / total_terremotos) * 100
    print(porcentaje)
else:
    print("No hay ningun terremoto acompañado de tsunamis en Indonesia")
```

>3. Calcular las estadísticas resumidas de los terremotos en Nevada.

```python
terremotos_nevada= terremotos[(terremotos["type"] == "earthquake") & (terremotos["place"] == "Nevada")]

media = terremotos_nevada.mean()

mediana = terremotos_nevada.median()

moda = terremotos_nevada.mode()

desviacion_estandar= terremotos_nevada.std()

valor_max = terremotos_nevada.max()

valor_min = terremotos_nevada.min()

cuantiles = terremotos_nevada.quantile([0.25,0.5,0.75])

suma= terremotos_nevada.cumsum()
```

```python
cinturon_fuego = ["Alaska", "Antarctic", "Bolivia", "California", "Canada", "Chile",
                    "Costa Rica", "Ecuador", "Fiji", "Guatemala", "Indonesia", "Japan",
                    "Kermadec Islands", "Mexico", "New Zealand", "Peru", "Philippines",
                    "Russia", "Taiwan", "Tonga", "Washington"]


terremotos["cinturon de fuego"] = terremotos["place"].apply(lambda x: x in cinturon_fuego)

terremotos.head(2)
```

>5. Calcula el número de terremotos en los lugares del Cinturón de Fuego y el número fuera de ellos.

```python
trues = len(terremotos[terremotos["cinturon de fuego"]== True])
print(trues)

falses = len(terremotos[terremotos["cinturon de fuego"]== False])
print(falses)
```
