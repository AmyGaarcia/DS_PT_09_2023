```python
#que elementos tiene la funcion y como lo harias 

cardinalidad= pd.DataFrame({
    "Cardinalidad": df.nunique(),
    "Porcentaje": (df.nunique() / len(df) * 100).round(2),
    "Tipo de dato": df.dtypes,
    "Valores Unicos": [df[columns].unique() for columns in df.columns], 

})

cardinalidad
```

```python
def cardinalidad(df):
        columnas= pd.DataFrame({
        "Cardinalidad":df.nunique(),
        "Porcentaje":(df.nunique()/len(df) *100).round(2),
        "Tipo de dato":df.dtypes,
        "Valores Unicos":[df[columns].unique() for columns in df.columns]})

        columnas["Tipo de variable"]=["Cualitativa" if df[col].dtype=="object" else "cuantitativa" for col in df.columns]
      print 

# cualitativa: nominal, ordinal
# cuantitativa: discreta, countinua


        return columnas





resultados = cardinalidad(df)
resultados
```
